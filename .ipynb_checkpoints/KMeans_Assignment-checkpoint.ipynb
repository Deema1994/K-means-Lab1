# EDA

## Create a scatterplot of Grad.Rate versus Room.Board where the points are colored by the Private column. 

```python
import seaborn as sns
plt.figure(figsize=(15,8))
sns.set_style("whitegrid")
sns.scatterplot(data = df , x = "Room.Board" , y = "Grad.Rate"   , hue="Private" ,color = "Private" );

```

## Create a scatterplot of F.Undergrad versus Outstate where the points are colored by the Private column.

```python
plt.figure(figsize=(15,8))
sns.set_style("whitegrid")
sns.scatterplot(data = df , x = "Outstate" , y = "F.Undergrad"   , hue="Private" ,color = "Private" , alpha=0.3);
```

## Use sns.FacetGrid to create a stacked histogram showing Out of State Tuition based on the Private column. 

```python
sns.set_style("darkgrid")

g = sns.FacetGrid(df, hue="Private", height=6 , aspect=2)
g.map(sns.histplot, 'Outstate', stat="count", multiple="stack", element="bars",  alpha=0.3);

```

## Use sns.FacetGrid to create a stacked histogram showing Grad.Rate based on the Private column. 

```python
sns.set_style("darkgrid")

g = sns.FacetGrid(df, hue="Private", height=6 , aspect=2)
g.map(sns.histplot, 'Grad.Rate', stat="count", multiple="stack", element="bars",  alpha=0.3);
```

## K Means Model

```python
from sklearn.cluster import KMeans
```

```python
kmeans = KMeans(n_clusters=2)
```

## Fit the model to all the data except for the Private label.

```python
X = df.drop('Private',axis=1)
```

```python
y_kmeans = kmeans.fit_predict(X)
```

## What are the cluster center vectors?

```python
kmeans.cluster_centers_
```

## Create a new column for df called 'Cluster', which is a 1 for a Private school, and a 0 for a public school.

```python
df['Cluster'] = np.where(df['Private'] == "Yes", 1, 0)
y = df['Cluster']
df.head()
```

## Create a confusion matrix and classification report to see how well the Kmeans clustering worked without being given any labels.

```python
from sklearn.metrics import confusion_matrix
confusionMatrix = confusion_matrix(y,y_kmeans)
print(confusionMatrix)
```

```python
from sklearn.metrics import classification_report
classificationReport = classification_report(y,y_kmeans)
print(classificationReport)
```
